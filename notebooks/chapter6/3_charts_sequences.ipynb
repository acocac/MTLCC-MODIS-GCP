```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib import gridspec
import os
import numpy as np
import glob
```

```python
indir = r'E:\acocac\research\AMZ\trajectories\pngs'
outdir = r'E:\acocac\research\AMZ\img\c6\lcdistribution'

seq_pngs = glob.glob(os.path.join(indir,'*_simple.png'))
```

```python
nrow = 4
ncol = 1
fig, axs = plt.subplots(nrow, ncol, figsize=(22,16),dpi=150)
#fig.tight_layout() # Or equivalently,  "plt.tight_layout()"
plt.subplots_adjust(hspace = 0)

ymin = 2015
ymax = 2018

sYear = 11
for i, ax in enumerate(fig.axes):
    
    target = mpimg.imread(seq_pngs[i+sYear])
    ax.imshow(target)
    
    ax.annotate('20'+str(i+sYear+4).zfill(2),(-0.02,0.5), xycoords='axes fraction', ha='center', va='center', fontsize='x-large', rotation=90, fontweight='bold')
    ax.axis('off')
    
plt.show()

fig.savefig(os.path.join(outdir,'charts_{}to{}.png'.format(ymin,ymax)), bbox_inches='tight',pad_inches = 0, dpi=150)
plt.close()
```
