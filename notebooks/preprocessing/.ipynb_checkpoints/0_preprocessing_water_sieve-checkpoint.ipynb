```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib import gridspec
import os
import numpy as np
```

```python
import rasterio
import matplotlib 
import matplotlib.pyplot as plt
from collections import Counter
import fiona
from rasterio.mask import mask
import geopandas as gpd
```

```python
def read_file(raster, shape):
    with rasterio.open(raster) as src:
        out_image, out_transform = mask(src, shape, crop=True)
        out_image = out_image.squeeze()
        return out_image
    
def bin_terra(arr):
    arr[arr <= 1] = 0
    arr[arr == 2019] = 0
    arr[arr != 0] = 1
    return(arr)
```

```python
indir = os.path.join(r'F:\acoca\research\gee\dataset\AMZ\implementation')
outdir = r'E:\acocac\research\AMZ\img\c6\preprocessing'

aoi_file = os.path.join(indir, 'aoi', 'amazon_raisg.shp')
aoi_viz = gpd.read_file(aoi_file)

# read files
with fiona.open(aoi_file, "r") as shapefile:
    aoi = [feature["geometry"] for feature in shapefile]
    
terra_raw = bin_terra(read_file(os.path.join(indir, 'terrai', 'AMZ_decrease_2004_01_01_to_2019_06_10.tif'), aoi))
water = read_file(os.path.join(indir, 'watermask', 'watermask.tif'), aoi)
terra_clump = read_file(os.path.join(indir, 'terrai_sieve', 'AMZ_sieve_terrai_2004_2018_sieve2.tif'), aoi)
```

```python
new_msk = (water != 0)  # working mask water and year
terra_water = np.ma.MaskedArray(terra_raw, mask=new_msk)
```

```python
colormap_terra = matplotlib.colors.LinearSegmentedColormap.from_list(range(2), ['#ffffff','#fa0000'])
colormap_water = matplotlib.colors.LinearSegmentedColormap.from_list(range(2), ['#ffffff','#0032c8'])
```

```python
colors = ['#0032c8','#fa0000']
labels = ['Water','2004-2019 Terra-i loss']

f = lambda m,c: plt.plot([],[],marker=m, color=c, ls="none")[0]
handles = [f("s", colors[i]) for i in range(len(colors))]
# legend = plt.legend(handles, labels, loc=5, framealpha=1, frameon=False)
legend = plt.legend(bbox_to_anchor=(0.5, -1.5), handletextpad=0.001, frameon=False, loc='lower center', borderaxespad=0., labels = labels, ncol=3, fontsize='xx-large')

def export_legend(legend, filename=os.path.join(outdir,"legend_preprocessingterrai.png"), expand=[-5,-5,5,5]):
    fig  = legend.figure
    fig.canvas.draw()
    bbox  = legend.get_window_extent()
    bbox = bbox.from_extents(*(bbox.extents + np.array(expand)))
    bbox = bbox.transformed(fig.dpi_scale_trans.inverted())
    fig.savefig(filename, dpi="figure", bbox_inches=bbox)

export_legend(legend)
plt.show()
```

```python
legend = mpimg.imread(os.path.join(outdir,'legend_preprocessingterrai.png'))

gs0 = gridspec.GridSpec(3, 4,  hspace=0.1, wspace=0.1, height_ratios= [0.2, 1])
arrays_l1 = [legend, water, terra_raw, terra_water, terra_clump]

fig = plt.figure(figsize=(16, 8), constrained_layout=False)
for i, arr in enumerate(arrays_l1):
    if i == 0:
        ax00 = fig.add_subplot(gs0[0, :])
        ax00.imshow(arr)
    elif i == 1:
        ax00 = fig.add_subplot(gs0[1,i-1])
        ax00.set_title('\n Watermask \n', fontsize='x-large')
        ax00.imshow(arr, cmap=colormap_water, interpolation='none', vmin=0, vmax=1, extent=aoi_viz.total_bounds[[0,2,1,3]])
        aoi_viz.plot(edgecolor='gray', linewidth=0.8, facecolor='None', ax=ax00, zorder=3)
    else:
        if i == 2:
            trat = '\n Raw'
            N = Counter(arr.flatten())[1.0]
        elif i == 3:
            trat = '\n mask 1 (water)'
            N = Counter(arr.compressed())[1.0]
        elif i == 4:
            trat = '\n mask 1 (water) + mask 2 (isolated)'
            N = Counter(arr.flatten())[1.0]
            
        ax00 = fig.add_subplot(gs0[1, i-1])
        ax00.set_title('Terra-i ' + trat + '\n (N=' + str(N) + ')', fontsize='x-large')
        ax00.imshow(arr, cmap=colormap_terra, interpolation='none', vmin=0, vmax=1, extent=aoi_viz.total_bounds[[0,2,1,3]])
        aoi_viz.plot(edgecolor='gray', linewidth=1, facecolor='None', ax=ax00, zorder=3)

    ax00.axis('off')
    
plt.show()

fig.savefig(os.path.join(outdir,'terrai_preprocessed.png'), bbox_inches='tight',pad_inches = 0, dpi=150)
plt.close()
```
