## Retrieve operations from the graph

````python
def get_op(name):
    return tf.get_default_graph().get_operation_by_name(name).outputs[0]

## get variables from tf.default_graph
iterator_handle_op = get_op("data_iterator_handle")
is_train_op = get_op("is_train")
global_step_op = get_op("global_step")
train_op = get_op("train_op")

# container to store operations to be obtained
query_map=dict()

#atrousdeep generation
query_map["x"]="input/reshaped/x"

query_map["convrnn_input"]="convrnn1/input"
query_map["convrnn_output"]="convrnn1/outputs"
query_map["convrnn_state"]="convrnn1/final_states"
    
query_map["targets"]="targets"
query_map["predictions"]="predictions"
query_map["prediction_scores"]="prediction_scores"
query_map["correctly_predicted"]="correctly_predicted"
````

## Inference - Execute the graph with data and store results in dict

````python
feed = {iterator_handle_op:data_handle,is_train_op:False}

operations = ops = [get_op(query_map[key]) for key in sorted(query_map.keys())]

# run a network session with the operations defined prviously
queried = sess.run(operations,feed_dict=feed)

# build a results dict
results=dict()
for key, array in zip(sorted(query_map.keys()),queried):
    results[key]=array
````

### Define the GRU cell
see implementation in ```utils/convrnn/cell.py```

````python
b,t,px,px,d_in = results["convrnn_input"].shape

def convolution(inputs,W,data_format):
    """wrapper around tf.nn.convolution with custom padding"""
    pad_h = int(W.get_shape()[0])/2
    pad_w = int(W.get_shape()[1])/2

    pad_h = int(float(pad_h))
    pad_w = int(float(pad_w))
    
    paddings = tf.constant([[0, 0], [pad_h,pad_h], [pad_w,pad_w], [0, 0]], dtype=tf.int32)
    
    inputs_padded = tf.pad(inputs, paddings, "REFLECT")

    return tf.nn.convolution(inputs_padded, W, 'VALID', data_format=data_format)

def layer_norm(inputs,beta,gamma):
    """taken from contrib tf.contrib.layers.layer_norm definition in
    tensorflow/contrib/layers/python/layers/layers.py
    """
    mean, variance = tf.nn.moments(inputs, [1,2,3], keep_dims=True)

    outputs = tf.nn.batch_normalization(
        inputs, mean, variance, offset=beta, scale=gamma,
        variance_epsilon=1e-12)
    return outputs

x = results["convrnn_input"][:,0]

scope_gates = "convrnn1/bidirectional_rnn/fw/conv_gru_cell/gates"

weights_gates=[]
weights_gates.append(get_op(scope_gates+"/kernel"))
weights_gates.append(get_op(scope_gates+"/LayerNorm/beta"))
weights_gates.append(get_op(scope_gates+"/LayerNorm/gamma"))
weights_gates.append(get_op(scope_gates+"/LayerNorm_1/beta"))
weights_gates.append(get_op(scope_gates+"/LayerNorm_1/gamma"))

scope_candidate = "convrnn1/bidirectional_rnn/fw/conv_gru_cell/candidate"

weights_candidate=[]
weights_candidate.append(get_op(scope_candidate+"/kernel"))
weights_candidate.append(get_op(scope_candidate+"/LayerNorm/beta"))
weights_candidate.append(get_op(scope_candidate+"/LayerNorm/gamma"))

def gru(x,state,weights_gates,weights_candidate, activation=tf.nn.tanh):
    """Implementation modified from carlthome/tensorflow-convlstm-cell"""

    h = state
    X = x
    
    convfilters = h.shape[-1]
    
    #gates
    kernel, b_r, g_r, b_u, g_u = weights_gates

    x = tf.concat([X, h], axis=3).eval()

    y = convolution(x,kernel,data_format="NHWC").eval()
    
    r, u = tf.split(y, 2, axis=3)

    r = layer_norm(r,b_r,g_r)
    u = layer_norm(u,b_u,g_u)
    
    r, u = tf.sigmoid(r), tf.sigmoid(u)

    #candidate
    kernel, b_h, g_h = weights_candidate

    x = tf.concat([X, r * h], axis=3).eval()
    
    y = convolution(x,kernel,data_format="NHWC").eval()
    
     ##not working if y is normalised using TF function!
#     y = tf.contrib.layers.layer_norm(y)

    # working if y splitted as r and u gates
    y = tf.split(y, 1, axis=3)
    y = y[0]
    y = layer_norm(y,b_h,g_h)

    h = u * h + (1 - u) * activation(y)

    return h,r,u
````

### Perform Iteration on CPU

````python
inputs = results["convrnn_input"]

b,t,px,px,d = results["convrnn_output"].shape
convfilters = d/2

zero_state_op=tf.zeros(tf.TensorShape([b, px, px, convfilters]))
state=sess.run(zero_state_op)

rGate = []
uGate = []
statesh = []

# execute on cpu because ressource exhausted error on GPU
with tf.device('/cpu:0'):

    for time in range(0,t):
        print("time {}".format(time))

        h,r,u = gru(inputs[:,time], state, weights_gates, weights_candidate)

        rGate.append(r.eval())
        uGate.append(u.eval())
        statesh.append(h.eval())
                
uGate = np.stack(uGate,axis=1)
rGate = np.stack(rGate,axis=1)
statesh = np.stack(statesh,axis=1)
````

## Write Arrays to png files

folder structure
``
{outfolder}/sample{s}/time{t}/{d}_{name}.png
``

with `s` as sample in batch, `t` as time, `d` as number of feature map and `name` `iGate`,`jGate`,`oGate`, etc.

````python
from __future__ import print_function
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import numpy as np
from PIL import Image

def show_rgb(x, name=""):
    
    if len(x.shape)==5: # BTHWD
        max_b, max_t,_,_,max_d = x.shape
    elif len(x.shape)==4: # BHWD
        max_b,_,_,max_d = x.shape
    
    def norm(band):
        return (band - band.min()) / (band - band.min()).max()
    
    def _show_map_BTHWD(t,d,b): 
        plt.title("{name} RGB map {rd}-{gn}-{bl}, b={b}, t={t}".format(name=name,b=b,t=t,rd=d-1,gn=d,bl=d+1))
        plt.imshow(np.stack((norm(x[b,t,:,:,d-1]),norm(x[b,t,:,:,d]),norm(x[b,t,:,:,d+1])),axis=-1))
        
    def _show_map_BHWD(d,b): 
        plt.title("{name} RGB map {rd}-{gn}-{bl}, b={b}".format(name=name,b=b,rd=d-1,gn=d,bl=d+1))
        plt.imshow(np.stack((norm(x[b,:,:,d-1]),norm(x[b,:,:,d]),norm(x[b,:,:,d+1])),axis=-1))
        
    # both
    b_slider = widgets.IntSlider(description='batch',min=0,max=max_b-1,step=1,value=max_b/2)
    d_slider = widgets.IntSlider(description='band',min=1,max=max_d-2,step=1,value=max_d/2) 
        
    if len(x.shape)==5: # BTHWD
        t_slider = widgets.IntSlider(description='time',min=0,max=max_t-1,step=1,value=max_t/2)
        w = interactive(_show_map_BTHWD, t=t_slider, d=d_slider, b=b_slider)
        
    elif len(x.shape)==4: # BHWD
        w = interactive(_show_map_BHWD, d=d_slider, b=b_slider)
    
    w.layout.height = '800px'
    display(w)


def show_gray(x, name="",vmin=None, vmax=None):
    
    if len(x.shape)==5: # BTHWD
        max_b, max_t,_,_,max_d = x.shape
    elif len(x.shape)==4: # BHWD
        max_b,_,_,max_d = x.shape
    elif len(x.shape)==3: # BHW
        max_b,_,_ = x.shape
        
    def _show(x,title):
        plt.title(title)
        plt.imshow(x,vmax=vmax, vmin=vmin);
        plt.colorbar()
        
        
    def _show_map_BTHWD(t,d,b): 
        _show(x[b,t,:,:,d],"{name} feature map b={b}, t={t}, d={d}".format(name=name,b=b,t=t,d=d))
        
    def _show_map_BHWD(d,b): 
        _show(x[b,:,:,d],"{name} feature map b={b}, d={d}".format(name=name,b=b,d=d))
        
    def _show_map_BHW(b): 
        _show(x[b,:,:],"{name} feature map b={b}".format(name=name,b=b))
        
    # all
    b_slider = widgets.IntSlider(description='batch',min=0,max=max_b-1,step=1,value=max_b/2)
         
    if len(x.shape)==5: # BTHWD
        d_slider = widgets.IntSlider(description='band',min=0,max=max_d-1,step=1,value=max_d/2) 
        t_slider = widgets.IntSlider(description='time',min=0,max=max_t-1,step=1,value=max_t/2)
        w = interactive(_show_map_BTHWD, t=t_slider, d=d_slider, b=b_slider)
        
    elif len(x.shape)==4: # BHWD
        d_slider = widgets.IntSlider(description='band',min=0,max=max_d-1,step=1,value=max_d/2) 
        w = interactive(_show_map_BHWD, d=d_slider, b=b_slider)
        
    elif len(x.shape)==3: # BHW
        w = interactive(_show_map_BHW, b=b_slider)
    
    w.layout.height = '800px'
    display(w)
    
def show(x,name="",mode="RGB"):
    if mode=="RGB":
        show_rgb(x,name)
    elif mode=="gray":
        show_gray(x,name)
    
def norm_ptp(arr):
    return (arr-arr.min()) / (arr-arr.min()).max()


def norm_std(arr,stddev=1):
    arr -= arr.mean(axis=0).mean(axis=0)
    arr /= stddev*arr.std(axis=0).std(axis=0) # [-1,1]
    arr = (arr/2) + 0.5 # [0,1]
    arr = np.clip(arr*255,0,255) # [0,255]
    return arr.astype("uint8")

def norm_rgb(arr):
    # taken from QGIS mean +- 2 stddev over cloudfree image
    vmin = np.array([-0.0433,-0.0054,-0.0237])
    vmax = np.array([0.1756,0.1483,0.1057])

    arr-=vmin
    arr/=(vmax-vmin)

    return np.clip((arr*255),0,255).astype("uint8")

def write(arr,outfile):
    #norm_img = norm(arr)
    img = Image.fromarray(arr)
    img.save(outfile)

def dump3(array,name,outfolder,cmap="inferno",norm=norm_ptp):
    
    filenpath="{outfolder}/sample{s}/{name}/{d}.png"

    cmap = plt.get_cmap(cmap)

    # normalize over the entire array
    #array = norm(array)
    
    samples,h,w,depth = array.shape
    for s in range(samples):
        for d in range(depth):
            outfilepath = filenpath.format(outfolder=outfolder,s=s,name=name,d=d)

            if not os.path.exists(os.path.dirname(outfilepath)): 
                os.makedirs(os.path.dirname(outfilepath))
            arr = array[s,:,:,d]
            arr = cmap(arr)

            write((arr*255).astype('uint8'),outfilepath)

def dump(array,name,outfolder,cmap="inferno",norm=norm_ptp):
    
    filenpath="{outfolder}/sample{s}/time{t}/{d}_{name}.png"

    cmap = plt.get_cmap(cmap)

    # normalize over the entire array
    #array = norm(array)
    
    samples,times,h,w,depth = array.shape
    for s in range(samples):
        for t in range(times):
            for d in range(depth):
                outfilepath = filenpath.format(outfolder=outfolder,s=s,t=t,name=name,d=d)

                if not os.path.exists(os.path.dirname(outfilepath)): 
                    os.makedirs(os.path.dirname(outfilepath))
                arr = array[s,t,:,:,d]
                arr = cmap(arr)

                write((arr*255).astype('uint8'),outfilepath)
    
def dump_rgb(array,name,outfolder,stddev):
    
    filenpath="{outfolder}/sample{s}/time{t}_{name}.png"

    
    
    samples,times,h,w,depth = array.shape
    for s in range(samples):
        for t in range(times):
            outfilepath = filenpath.format(outfolder=outfolder,s=s,t=t,name=name)

            if not os.path.exists(os.path.dirname(outfilepath)): 
                os.makedirs(os.path.dirname(outfilepath))
            arr = array[s,t,:,:,0:3]
            
            arr = norm_std(arr,stddev=stddev)

            write(arr,outfilepath)
        
def dump_class(array,name,outfolder,cmap="Accent"):
    filenpath="{outfolder}/sample{s}/{name}.png"
    samples,h,w = array.shape

    array = array.astype(float) / 17
    
    cmap = plt.get_cmap(cmap)
    for s in range(samples):
        outfilepath = filenpath.format(outfolder=outfolder,s=s,name=name)


        arr = (cmap(array[s])*255).astype("uint8")
        write(arr,outfilepath)
````
